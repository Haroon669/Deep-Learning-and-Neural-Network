```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./Financial Distress.csv')
```

```python
df.head()
```

```python
## Check if Company is greater than -0.5 The it is Healthy Companies otherwise it is Finacial Distress
##.We are making Categorical Response Variable to determine Healthy and Financial Distress Companies
df['DF'] = pd.np.where(df['Financial Distress'] > -0.50,0,1)
df.head()
```

```python
df.info()
```

```python
df.shape   # 3672 Rows and 87 Columns
```

```python
df.describe()
```

```python
columns = ['Company','Time','x80','DF']
```

```python
# To make four Column Categorical Column We are changing thes type to Object
for i in columns:
    df[i] = df[i].astype(str)
```

```python
df.info()
```

```python
df.head()
```

```python
df=df.drop('Financial Distress',axis=1)  # To drop Finacial Distress Column as no longer need of this Column
```

```python
# Check if  Column contain any Null Values or Not
df.isnull().sum()
```

```python
df.shape[0]
```

```python
 # To check null on every Columns 
df.isnull().sum()/df.shape[0]*100  
```

```python
# We are going to drop our Response Variable FD From our Predictive Model.
X = df.drop('DF',axis=1)
X.head()
```

```python
Y = df['DF']
Y.head()
```

```python
# Let count the Count of Values.
# Non-Distress Companies denote 0 and Distress Companies denote 1
Y.value_counts()
```

```python
X.info()
```

```python
# To make X as Categorical Data get_Dummies() method is applied.One Hot Enconding is applied to it.
X = pd.get_dummies(X)
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(whiten=True)
pca.fit(X_train)
```

```python
variance = pd.DataFrame(pca.explained_variance_ratio_)
variance
```

```python
np.cumsum(pca.explained_variance_ratio_)
#This will tell How many Components pca need to learn to reach 1 or 100 % Learning so It learned all feature on 15.
```

```python
#so we are reducing component ratio
pca = PCA(n_components=15,whiten=True)
pca_X_train = pca.fit_transform(X_train)
pca_X_test = pca.transform(X_test)
```

```python
#Now Scaling the Data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(pca_X_train)
X_test = sc.transform(pca_X_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1)
```

```python
mlp.fit(X_train,Y_train)
```

```python
y_predict = mlp.predict(X_test)
```

```python
# lets check Test data and Predicted Value of Test Data
from sklearn.metrics import accuracy_score 
accuracy_score(Y_test,y_predict)
# It have given 96% Accuracy 
```

```python
# Confusion Matrix is used to find Accuracy
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_predict)
```
